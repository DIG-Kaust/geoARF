## Carrier frequency space

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker, colors

plt.style.use('ggplot')

L = np.linspace(5,75,100)
r = np.linspace(1,10,100)

LL, rr = np.meshgrid(L, r)
f = (LL*1500)/(rr**2)

# Mask values of f above 5000
f_masked = np.ma.masked_where(f > 5000, f)

plt.figure(figsize=(15, 10))

# Specify the contour levels
lev_exp = np.linspace(2, 5, 16)
levs = np.power(10, lev_exp)

# Create a filled contour plot of f in log scale, using the masked array
cs = plt.contourf(LL, rr, f, levs, norm=colors.LogNorm(), cmap='turbo')

ass = plt.contour(LL, rr, f, levs, colors='black', alpha=0.1)

# Create a colorbar for the contour plot with log-spaced ticks
cbar = plt.colorbar(cs, label='Frequency [Hz]')
cbar.locator = ticker.LogLocator() # Set locator to logarithmically spaced intervals
cbar.update_ticks() # Update colorbar ticks

plt.xlabel('Focus depth [m]')
plt.ylabel('Transducer radius [m]')
plt.title('Frequency space')
plt.show()
```

## Resolution space

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker, colors

plt.style.use('ggplot')

v = np.linspace(50, 400, 100)
fm = np.linspace(1, 150, 100)

vv, ffmm = np.meshgrid(v, fm)

res = vv/(ffmm*4)

plt.figure(figsize=(15, 10))

# Specify the contour levels
lev_exp = np.linspace(-1, 2, 16)
levs = np.power(10, lev_exp)

# Create a filled contour plot of f in log scale, using the masked array
cs = plt.contourf(vv, ffmm, res, levs, norm=colors.LogNorm(), cmap='turbo')

ass = plt.contour(vv, ffmm, res, levs, colors='black', alpha=0.1)

plt.plot(v, v/(0.5*4), lw=4, c='k', ls='--')

# Create a colorbar for the contour plot with log-spaced ticks
cbar = plt.colorbar(cs, label='Resolution [m]')
cbar.locator = ticker.LogLocator() # Set locator to logarithmically spaced intervals
cbar.update_ticks() # Update colorbar ticks

plt.xlabel('Seabed velocity [m/s]')
plt.ylabel('Modulating frequency [Hz]')
plt.suptitle('Resolution space')
plt.tight_layout()
plt.ylim(0, 150)
plt.show()
```
